```python
chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
```

```python
def appending():
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    Channel_name1 = soup.select('label.Filter_label_DUyBa')
    for i in Channel_name1:
        i = str(i)
        i = i.split('</svg>')[1]
        i = i.split('<')[0]
        temp_list.append(i)
```

```python
url = 'https://shoppinglive.naver.com/categories/dc:9'
temp_list = []
temp_list_search = []
driver.get(url)
time.sleep(3)

# 채널클릭
buttun2 = driver.find_element(By.XPATH, '//*[@id="channel"]')
buttun2.click()
time.sleep(1)

# ㄱ ~ ㅎ 추출
for i in range(1,15):
    try:
        buttun_1 = driver.find_element(By.XPATH, f'//*[@id="wa_filter_depth3_tab_{i}"]')
        buttun_1.click()
        time.sleep(1)
        appending()
    except:
        continue
time.sleep(1)

# 영어 채널 클릭.
buttun3 = driver.find_element(By.XPATH, '//*[@id="wa_filter_depth2_tab_1"]')
buttun3.click()
time.sleep(1)

# A~Z
for i in range(1,27):
    try:
        buttun_4 = driver.find_element(By.XPATH, f'//*[@id="wa_filter_depth3_tab_{i}"]')
        buttun_4.click()
        time.sleep(1)
        appending()
    except:
        continue

```
