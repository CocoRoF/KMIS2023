```python
Path_1 = '/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/nonTour_df'
Path_2 = '/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Tour_only_df'
Mall_super_list = os.listdir(Path_1)
Mall_tour_list = os.listdir(Path_2)
Mall_super_list.remove('Complete_list.pkl')
Mall_super_list.remove('Check_list.pkl')
Mall_tour_list.remove('Complete_list.pkl')
Mall_tour_list.remove('Check_list.pkl')
```

```python
okt = Okt()
dataset = []

for mall in Mall_super_list:
  Video_Path = Path_1 + f'/{mall}'
  video_list = os.listdir(Video_Path)
  print(f'Start : {mall}')
  bar = progressbar.ProgressBar(maxval = len(video_list))
  for video in video_list:
    if 'csv' not in video:
      video_list.remove(video)
      print(f'remove : {video}')

  for video in video_list:
    if len(video) > 17:
      video_list.remove(video)
      print(f'remove : {video}')

  for k, video in enumerate(video_list):
    bar.update(k)
    try:
      video_path = Video_Path + f'/{video}'
      df = pd.read_csv(video_path)
      for i in df['chat']:
        words = okt.nouns(i)
        if len(words) > 0:
          dataset.append(words)
    except:
      print(f'Error Occur : {video_path}')
  bar.finish()

Total_df = pd.DataFrame(list(apriori(dataset, min_support = 0.0001)))
Total_df['length'] = Total_df['items'].apply(lambda x: len(x))
Total_df.to_csv('/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Keyword_Network/종합몰_Keyword_원본.csv', index=False)
Total_df.to_excel('/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Keyword_Network/종합몰_Keyword_원본.xlsx', index=False)

Total_df_2 = Total_df[(Total_df['length'] == 2) & (Total_df['support'] > 0.001)].sort_values(by = 'support', ascending = False)
Total_df_2.to_csv('/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Keyword_Network/종합몰_Keyword_2.csv', index=False)
Total_df_2.to_excel('/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Keyword_Network/종합몰_Keyword_2.xlsx', index=False)
```

```python
Total_df_2 = pd.read_csv('/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Keyword_Network/종합몰_Keyword_원본.csv')
Total_df_2 = Total_df_2[(Total_df_2['length'] == 2)].sort_values(by = 'support', ascending = False)
```

```python
Total_df_3 = Total_df_2[Total_df_2['length'] == 2]
```

```python
item_tuple_list = []
for i in Total_df_3['items']:
  a = i.split("'")
  item_1 = a[1]
  item_2 = a[3]
  if (len(item_1)>1) & (len(item_2)>1):
    item_tuple = (a[1], a[3])
    item_tuple_list.append(item_tuple)
```

```python
len(item_tuple_list)
```

```python
G = nx.Graph()
G.add_edges_from(item_tuple_list)
```

```python
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 2000 * (nsize - min(nsize)) / (max(nsize) - min(nsize))
```

```python
nx.write_gexf(G, "/content/drive/MyDrive/Project/SuperMall_test_2.gexf")
```

```python
# pos = nx.planar_layout(G)
plt.figure(figsize = (16,16)); plt.axis('off')
nx.draw_networkx(G, node_size=nsize, alpha=0.7)
plt.show()
```
