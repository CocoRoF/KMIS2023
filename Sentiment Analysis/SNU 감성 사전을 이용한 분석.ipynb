```python
!pip install konlpy
import pandas as pd
import os
from konlpy.tag import Okt

sentiword = '/content/drive/MyDrive/LiveCommerse/KnuSentiLex-master/KnuSentiLex-master/SentiWord_Dict.txt'

# 감성 사전 생성 함수
def load_dictionary():
    dictionary = pd.read_csv(sentiword, sep='\t', header=None, names=['word', 'score'])
    return dictionary

dictionary = load_dictionary()

okt = Okt()

def analyze_sentiment(text):
    words = okt.pos(text, stem=True, norm=True)

    word_scores = []

    for word, pos in words:
        if pos == 'Adjective' or pos == 'Verb':
            try:
                score = dictionary.loc[dictionary['word'] == word]['score'].values[0]
                word_scores.append(score)
            except:
                continue

    if len(word_scores) > 0:
        avg_score = sum(word_scores) / len(word_scores)
    else:
        avg_score = 0.0

    if avg_score > 0:
        sentiment = 'Positive'
    elif avg_score < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'

    return sentiment, avg_score

def Describe(df, save_df, Video_number):
  df['Sentiment'] = 0
  df['Score'] = 0
  df['chat'] = df['chat'].astype(str)

  for i in range(len(df)):
    df['Sentiment'][i] = analyze_sentiment(df['chat'][i])[0]
    df['Score'][i] = analyze_sentiment(df['chat'][i])[1]

    
  pattern = '\d{16,}'
  hello_pattern = '안녕|방가|반가|하이'
  question_pattern = '요\?|죠\?|용\?|까\?|여\?|염\?'

  drop_df = df[~(df['chat'].str.contains(pattern))]
  drop_df.reset_index(drop=True, inplace=True)
  drop_df

  len_chat = len(df)
  drop_chat = len(drop_df)
  drop_user = drop_df.name.nunique()
  drop_positive = len(drop_df[drop_df['Score']>0])
  drop_negative = len(drop_df[drop_df['Score']<0])
  drop_neutral = len(drop_df[drop_df['Score']==0])
  drop_average = drop_df.Score.sum() / (len(drop_df))
  drop_hello = len(drop_df[(drop_df['chat'].str.contains(hello_pattern))])
  drop_question = len(drop_df[(drop_df['chat'].str.contains(question_pattern))])

  empty = []
  temp_df = pd.DataFrame({'VideoNum' : empty, 'Total_Chat' : empty, 'Droped_Chat' : empty, 'User_Num' : empty, 'Positive' : empty, 'Negative': empty, 'Neutral' : empty, 'Average' : empty, 'Num_hello' : empty, 'Num_Question' : empty})
  temp_df['VideoNum'][0] = Video_number
  temp_df['Total_Chat'][0] = len_chat
  temp_df['Droped_Chat'][0] = drop_chat
  temp_df['User_Num'][0] = drop_user
  temp_df['Positive'][0] = drop_positive
  temp_df['Negative'][0] = drop_negative
  temp_df['Neutral'][0] = drop_neutral
  temp_df['Average'][0] = drop_average
  temp_df['Num_hello'][0] = drop_hello
  temp_df['Num_Question'][0] = drop_question

  save_df = pd.concat([save_df, temp_df])
```
