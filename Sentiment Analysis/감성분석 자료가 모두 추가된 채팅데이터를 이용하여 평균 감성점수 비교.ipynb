```python
Path_1 = '/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/nonTour_df'
Path_2 = '/content/drive/MyDrive/Project/LiveCommerse/FUCKUGOOGLE/Tour_only_df'
Mall_super_list = os.listdir(Path_1)
Mall_tour_list = os.listdir(Path_2)
Mall_super_list.remove('Complete_list.pkl')
Mall_super_list.remove('Check_list.pkl')
Mall_tour_list.remove('Complete_list.pkl')
Mall_tour_list.remove('Check_list.pkl')
```

```python
mapping_dict = dict({4:2, 3:1, 2:0, 1:-1, 0:-2})
```

```python
total_df = pd.DataFrame()
a=0
for i in Mall_super_list:
  Channel_name = i.split('_')[0]
  Channel_num = i.split('_')[1]

  Video_Path = Path_1 + f'/{i}'
  video_list = os.listdir(Video_Path)

  for j in video_list:
    if 'csv' not in j:
      video_list.remove(j)
      print(f'remove : {j}')

  for k in video_list:
    if len(k) > 17:
      video_list.remove(k)
      print(f'remove : {k}')

  for video in video_list:
    video_num = video.split('_')[1]
    video_num = video_num.split('.')[0]
    try:
      video_path = Video_Path + f'/{video}'
      df = pd.read_csv(video_path)
      temp_df = df[['chat', 'Score', 'GPT_Sentiment_Score_1','KoBERT_Sentiment_Score_1']]
      temp_df['KoBERT_Sentiment_Score_2'] = temp_df['KoBERT_Sentiment_Score_1'].map(lambda x: mapping_dict.get(x))
      Dict_Score = temp_df.Score.sum() / len(temp_df)
      GPT_Score = temp_df.GPT_Sentiment_Score_1.sum() / len(temp_df)
      KoBERT_Score = temp_df.KoBERT_Sentiment_Score_2.sum() / len(temp_df)

      total_df.loc[a, 'Channel'] = Channel_name
      total_df.loc[a, 'Channel_Number'] = Channel_num
      total_df.loc[a, 'Video_Number'] = video_num

      total_df.loc[a, 'Dict_Score_Average'] = Dict_Score
      total_df.loc[a, 'GPT_Score_Average'] = GPT_Score
      total_df.loc[a, 'KoBERT_Score_Average'] = KoBERT_Score

      a = a+1
    except:
      total_df.loc[a, 'Channel'] = Channel_name
      total_df.loc[a, 'Channel_Number'] = Channel_num
      total_df.loc[a, 'Video_Number'] = video_num

      total_df.loc[a, 'Dict_Score_Average'] = 0
      total_df.loc[a, 'GPT_Score_Average'] = 0
      total_df.loc[a, 'KoBERT_Score_Average'] = 0

      a = a+1
total_df.to_excel('/content/drive/MyDrive/Project/LiveCommerse/OpenAI/종합몰_감성분석_정리자료.xlsx')
```
