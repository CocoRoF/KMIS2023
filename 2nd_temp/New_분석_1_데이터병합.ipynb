```python
import pandas as pd
```

```python
data_1 = pd.read_excel('G:\내 드라이브\Project\LiveCommerse\FUCKUGOOGLE\종합몰_g-DMR\종합몰_최종_수정불가.xlsx')
data_2 = pd.read_excel('G:\내 드라이브\Project\LiveCommerse\FUCKUGOOGLE\투어전문몰_g-DMR\투어전문몰_최종_수정불가.xlsx')
```

```python
data_1['TourOnly'] = 0
data_2['TourOnly'] = 1
total_data = pd.concat([data_1, data_2])
```

```python
total_data = total_data[total_data.Channel_Total_Chat > 0]
total_data.info()
```

```python
total_data.columns = ['Vdo_num', 'Name', 'Channer_num', 'Sub',
       'Total_Video', 'Total_Chat', 'Event_Chat',
       'No_Event_Chat', 'Chat_User', 'Dict_Positive',
       'Dict_Negative', 'Dict_Neutral',
       'Dict_Score_Average', 'GPT_Score_Average',
       'KoBERT_Score_Average', 'Hello_Chat', 'Question_Chat',
       'Chat_Num_per_user', 'words', 'Tour_only']
```

```python
total_data.describe()
```

```python
df_1 = pd.read_excel('G:\내 드라이브\Project\LiveCommerse\OpenAI\종합몰_감성분석_정리자료_최종_2번째파일_2.xlsx')
df_2 = pd.read_excel('G:\내 드라이브\Project\LiveCommerse\OpenAI\투어전문몰_감성분석_정리자료_최종_2번째파일.xlsx')
```

```python
df_1 = df_1[['Video_Number', 'GPT_Score_Positive', 'GPT_Score_Negative', 'GPT_Score_Neutral',
             'KoBERT_Score_Positive', 'KoBERT_Score_Negative', 'KoBERT_Score_Neutral']]
df_1.columns = ['Vdo_num', 'GPT_Positive', 'GPT_Negative', 'GPT_Neutral',
             'KoBERT_Positive', 'KoBERT_Negative', 'KoBERT_Neutral']
df_1.dropna(inplace=True)
```

```python
df_2 = df_2[['Video_Number', 'GPT_Score_Positive', 'GPT_Score_Negative', 'GPT_Score_Neutral',
             'KoBERT_Score_Positive', 'KoBERT_Score_Negative', 'KoBERT_Score_Neutral']]
df_2.columns = ['Vdo_num', 'GPT_Positive', 'GPT_Negative', 'GPT_Neutral',
             'KoBERT_Positive', 'KoBERT_Negative', 'KoBERT_Neutral']
df_2.dropna(inplace=True)
```

```python
df_3 = pd.concat([df_1, df_2])
```

```python
Data_Total = pd.merge(total_data, df_3, on='Vdo_num', how='inner')
```

```python
Data_Total['Event_Rate'] = Data_Total['Event_Chat'] / Data_Total['Total_Chat']
Data_Total['Hello_Rate'] = Data_Total['Hello_Chat'] / Data_Total['Total_Chat']
Data_Total['Question_Rate'] = Data_Total['Question_Chat'] / Data_Total['Total_Chat']
Data_Total['Dict_Positive_Rate'] = Data_Total['Dict_Positive'] / Data_Total['Total_Chat']
Data_Total['Dict_Negative_Rate'] = Data_Total['Dict_Negative'] / Data_Total['Total_Chat']
Data_Total['Dict_Neutral_Rate'] = Data_Total['Dict_Neutral'] / Data_Total['Total_Chat']
Data_Total['GPT_Positive_Rate'] = Data_Total['GPT_Positive'] / Data_Total['Total_Chat']
Data_Total['GPT_Negative_Rate'] = Data_Total['GPT_Negative'] / Data_Total['Total_Chat']
Data_Total['GPT_Neutral_Rate'] = Data_Total['GPT_Neutral'] / Data_Total['Total_Chat']
Data_Total['KoBERT_Positive_Rate'] = Data_Total['KoBERT_Positive'] / Data_Total['Total_Chat']
Data_Total['KoBERT_Negative_Rate'] = Data_Total['KoBERT_Negative'] / Data_Total['Total_Chat']
Data_Total['KoBERT_Neutral_Rate'] = Data_Total['KoBERT_Neutral'] / Data_Total['Total_Chat']
Data_Total['Event_Rate'] = Data_Total['Event_Chat'] / Data_Total['Total_Chat']
Data_Total['Event_Rate'] = Data_Total['Event_Chat'] / Data_Total['Total_Chat']
Data_Total['Event_Rate'] = Data_Total['Event_Chat'] / Data_Total['Total_Chat']
```

```python
Data_Total['Norm_Chat'] = Data_Total['Total_Chat'] - Data_Total['Event_Chat'] - Data_Total['Question_Chat'] - Data_Total['Hello_Chat']
```

```python
Data_Total = Data_Total[['Vdo_num', 'Name', 'Channer_num', 'Sub', 'Total_Video', 'Total_Chat', 'Chat_User', 'Chat_Num_per_user',
       'Norm_Chat', 'Event_Chat',  'Hello_Chat', 'Question_Chat', 'No_Event_Chat', 'Event_Rate', 'Hello_Rate','Question_Rate', 
       'Dict_Positive', 'Dict_Negative', 'Dict_Neutral', 'Dict_Positive_Rate', 'Dict_Negative_Rate', 'Dict_Neutral_Rate','Dict_Score_Average',
       'GPT_Positive', 'GPT_Negative', 'GPT_Neutral', 'GPT_Positive_Rate', 'GPT_Negative_Rate', 'GPT_Neutral_Rate','GPT_Score_Average', 
       'KoBERT_Positive', 'KoBERT_Negative', 'KoBERT_Neutral','KoBERT_Positive_Rate', 'KoBERT_Negative_Rate', 'KoBERT_Neutral_Rate','KoBERT_Score_Average', 
       'Tour_only', 'words']]
```

```python
Data_Total.to_csv('G:\내 드라이브\Project\LiveCommerse\Python_Code\Re_zero부터시작하는분석\Total_Data.csv', index=False)
Data_Total.to_excel('G:\내 드라이브\Project\LiveCommerse\Python_Code\Re_zero부터시작하는분석\Total_Data.xlsx', index=False)
```
